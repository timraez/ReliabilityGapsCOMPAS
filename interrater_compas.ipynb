# Interrater Reliability Gaps on COMPAS: Experiments

## Setup and train logistic regression models on COMPAS dataset:

```python
import numpy as np
import warnings

import train_eval as te
import experiments as exps

compas = {
    'tag': 'compas',
    'name': 'ProPublica recidivism',
    'filename': 'data/propublica-recidivism_numerical-binsensitive_slim.csv',
    'sensitive_attribute': 'race',
    'target': 'two_year_recid',
    'categorical_attributes': ['c_charge_degree_M', 'c_charge_degree_F', 'sex'],
    'numerical_attributes': ['age', 'juv_fel_count', 'juv_misd_count', 'juv_other_count', 'priors_count'],
    'minima_numerical_attributes': {'age': 18, 'juv_fel_count': 0, 'juv_misd_count': 0, 'juv_other_count': 0, 'priors_count': 0}
}

np.random.seed(0)

warnings.filterwarnings(action='ignore', category=RuntimeWarning)
# this ignores warnings for the computation of IRR statistics in the case of zero noise.

trained_models_folds = te.fit_model_data(compas)
```

### Explanation of 'parameters experiments' dictionary:

- 'probabilities': Choose array with levels of noise / probablities of disturbance. Below the interval [0, 0.3] with increments of 0.01 is used.
- 'grouped': Choose whether noise is applied group-wise or not. Values: 'Y' = yes, 'N' = no.
- 'num_minima': Choose whether minima for numerical attributes are used. Values: 'Y' = yes, 'N' = no.
- 'feature_set': Choose set of features to be perturbed. Values: 'all features', 'numerical features', 'categorical features'
- 'variance_list': array with variances of normal perturbation of numerical features Values: numerical array or None.

## Perturbation of all features, variance = 1, 5, 10:

```python
parameters_experiments_1 = {
    'probabilities': np.arange(0, 0.301, 0.01),
    'grouped': 'Y', 
    'num_minima': 'N',
    'feature_set': 'all',
    'variance_list': [1, 5, 10]
}

exps.experiments(compas, parameters_experiments_1, trained_models_folds)
```

```python
parameters_experiments_2 = {
    'probabilities': np.arange(0, 0.301, 0.01),
    'grouped': 'Y', 
    'num_minima': 'Y',
    'feature_set': 'all',
    'variance_list': [1, 5, 10]
}

exps.experiments(compas, parameters_experiments_2, trained_models_folds)
```

## Perturbation of numerical features only, variance = 1, 5, 10:
